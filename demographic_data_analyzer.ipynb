```python
# Question 4 what percentage of people with advanced education (Bachelors, Masters, or Doctorate) make more than 50K?
#df[(df["education"].isin(["Bachelors","Masters","Doctorate"]))] & df["salary" == ">50"]
#(df["salary" == ">50"].sum()/df[(df["education"].isin(["Bachelors","Masters","Doctorate"]))].sum())*100
(df[(df["education"].isin(["Bachelors", "Masters", "Doctorate"])) & (df["salary"] == ">50K")].shape[0]/df[df["education"].isin(["Bachelors", "Masters", "Doctorate"])].shape[0]) * 100
```

```python
# Question 5 What percentage of people without advanced education make more than 50K?
#please note i considered anything past 12th as advanced education which are "Bachelors", "Masters", "Doctorate","Some-college
df[(df["education"].isin(["Bachelors", "Masters", "Doctorate","Some-college"])) & (df["salary"] == ">50K")].shape[0]/df[df["education"].isin(["Bachelors", "Masters", "Doctorate","Some-college"])].shape[0]*100
```

```python
# Question 6 What is the minimum number of hours a person works per week?
df["hours-per-week"].min()
```

```python

# Question 7 What percentage of the people who work the minimum number of hours per week have a salary of more than 50K?
df[(df["hours-per-week"] == 1) & (df["salary"] == ">50K")].shape[0] / df[df["hours-per-week"] == 1].shape[0]*100
```

```python
# Question 8 What country has the highest percentage of people that earn >50K and what is that percentage



# Step 1: Group data by country and count people
high_earners_per_country = df[df["salary"] == ">50K"].groupby("native-country")["salary"].count()
total_people_per_country = df.groupby("native-country")["salary"].count()

# Step 2: Compute the percentage of high earners per country
percentage_high_earners = (high_earners_per_country / total_people_per_country) * 100

# Step 3: Find the country with the highest percentage
highest_percentage_country = percentage_high_earners.idxmax()  # Country with max percentage
highest_percentage = percentage_high_earners.max()  # Max percentage

# Print result
print(f"The country with the highest percentage of people earning >50K is {highest_percentage_country} with {highest_percentage:.2f}%")
```

```python
# Question 9 Identify the most popular occupation for those who earn >50K in India.


# Step 1: Filter for people in India earning >50K
india_high_earners = df[(df["native-country"] == "India") & (df["salary"] == ">50K")]

# Step 2: Find the most common occupation
most_common_occupation = india_high_earners["occupation"].value_counts().idxmax()

# Print result
print(f"The most popular occupation for those earning >50K in India is: {most_common_occupation}")
```
